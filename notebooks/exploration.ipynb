# 1. Imports & Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Load Data

```python
# Adjust path if needed
df = pd.read_csv("../data/raw/ridership.csv")
df.head()

```

# 3. Basic Info

```python
# Basic EDA
df.info()
df.describe()
```

# 4. Preprocessing

```python
df["date"] = pd.to_datetime(df["date"])
df["day_of_year"] = df["date"].dt.dayofyear
df = df.dropna()
```

# 5. Exploratory Plots

```python
# Ridership trend over time
plt.figure(figsize=(10,5))
sns.lineplot(x="date", y="passengers", data=df)
plt.title("Ridership Over Time")
plt.show()

# Delay distribution
plt.figure(figsize=(8,4))
sns.histplot(df["delay_minutes"], bins=20, kde=True)
plt.title("Delay Minutes Distribution")
plt.show()

# Ridership by line
plt.figure(figsize=(8,4))
sns.boxplot(x="line", y="passengers", data=df)
plt.title("Ridership by Line")
plt.show()
```

# 6. Correlation Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
